```python
# Library import
import tensorflow 
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
(x_train, y_train),(x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape, y_train.shape,"******"  ,x_test.shape, y_test.shape
```

```python
plt.imshow(x_train[0])
```

```python
x_train[0]
```

```python
y_train
```

```python
class_labels = ["Zero", "One", "Two", "Three" ,"Four" ,"Five" , "Six" , "Seven" , "Eight" ,"Nine"]
```

```python
plt.figure(figsize=(25,18))

j=1
for i in np.random.randint(0, 1000,25):
  plt.subplot(5,5,j);j+=1
  plt.imshow(x_train[i])
  plt.axis("off")
  plt.title("{} ".format(class_labels[y_train[i]],[y_train[i]]))
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation="relu"))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=35, validation_split=0.2)
```

```python
y_prob=model.predict(x_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
plt.figure(figsize=(25,18))

j=1
for i in np.random.randint(0, 1000,25):
  plt.subplot(5,5,j);j+=1
  plt.imshow(x_train[i])
  plt.axis("off")
  plt.title("Actual={} \nPredicted={}".format(class_labels[y_test[i]],y_test[i],[y_train[i]], class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
```

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(16,9))
y_pred_labels = [np.argmax(label) for label in y_pred] 
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt="d", xticklabels=class_labels, yticklabels=class_labels)
```
